# One Shot Learning

PES2UG23CS928

```python
from transformers import pipeline
```

```python
classifier = pipeline(
    "text-generation",
    model="google/flan-t5-small"
)
```

```python
prompt = """
Classify the sentiment of the feedback.

Example:
Feedback: "The class was very interesting"
Sentiment: Positive

Now classify:
Feedback: "The lab sessions were boring"
Sentiment:
"""
```

```python
prompt2 = """
Classify student performance.

Examples:
Marks: 93
Grade: S

Marks: 66
Grade: B

Now classify:
Marks: 75
Grade:
"""
```

```python
result1 = classifier(prompt, max_new_tokens=10)
print(result1[0]["generated_text"])
```

```python
result2 = classifier(prompt2, max_new_tokens=10)
print(result2[0]["generated_text"])
```

# Few Shot Learrning

PES2UG23CS928

```python
from transformers import pipeline
```

```python
classifier = pipeline(
    "text-generation",
    model="google/flan-t5-small"
)
```

```python
prompt = """
Classify the sentiment of the feedback.

Examples:
Feedback: "The class was very interesting"
Sentiment: Positive

Feedback: "The professor explained concepts well"
Sentiment: Positive

Feedback: "The lab sessions were boring"
Sentiment: Negative

Now classify:
Feedback: "The lab sessions were confusing"
Sentiment:
"""
```

```python
prompt2 = """
Classify student performance.

Examples:
Marks: 93
Grade: S

Marks: 66
Grade: B

Now classify:
Marks: 75
Grade:
"""
```

```python
result1 = classifier(prompt, max_new_tokens=10)
print(result1[0]["generated_text"])
```

```python
result2 = classifier(prompt2, max_new_tokens=10)
print(result2[0]["generated_text"])
```
